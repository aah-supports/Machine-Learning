# Correction - Descente de gradient à la main

On reprend le même exemple avec `x = 50` et `y = 100`.

L'objectif est de voir comment la descente de gradient pousse le poids vers la bonne valeur.

````python
import matplotlib.pyplot as plt
import numpy as np

x = 50
y = 100

def cost(w):
    return (y - x * w) ** 2

def gradient(w):
    return -100 * (100 - 50 * w)

weights = [0, 1, 1.5, 2, 2.5]
````

## 1. Valeurs du coût

On calcule `J(w)` pour plusieurs valeurs de `w`.

````python
for w in weights:
    print(f'w = {w} -> J(w) = {cost(w):.0f}')
````

## 2. Lecture des résultats

On obtient :

- `w = 0` → `10000`
- `w = 1` → `2500`
- `w = 1.5` → `625`
- `w = 2` → `0`
- `w = 2.5` → `625`

Le minimum est bien atteint en `w = 2`.

````python
ws = np.linspace(-0.5, 3.0, 200)
js = [cost(w) for w in ws]

plt.figure(figsize=(7, 4))
plt.plot(ws, js)
plt.axvline(2, color='green', linestyle='--', label='minimum')
plt.xlabel('w')
plt.ylabel('J(w)')
plt.title('Fonction de coût')
plt.legend()
plt.grid(alpha=0.2)
plt.show()
````

## 3. Dérivée

La pente vaut :

```text
J'(w) = -100(100 - 50w)
```

À `w = 1` :

```text
J'(1) = -5000
```

Le signe négatif indique qu'il faut augmenter `w`.

````python
print(gradient(1))
print('La pente est négative, donc on augmente w.')
````

## 4. Mise à jour

Avec `η = 0.0001` :

````python
eta = 0.0001
w = 1
w_new = w - eta * gradient(w)
print(w_new)
````

## 5. Plusieurs itérations

On applique la même règle plusieurs fois.

````python
w = 1
for _ in range(5):
    w = w - eta * gradient(w)
    print(round(w, 6))
````

## 6. Ce qu'il faut dire à l'oral

- `J(w)` mesure l'erreur.
- La dérivée donne la direction de la pente.
- On bouge dans le sens opposé à la pente pour faire baisser l'erreur.
- C'est exactement le mécanisme utilisé par les réseaux de neurones, avec beaucoup plus de poids.
